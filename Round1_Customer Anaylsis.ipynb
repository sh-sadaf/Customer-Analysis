```python
#Standardize header names.
#1.change columns names to lower case
file1.columns = file1.columns.str.lower()
```

```python
file2.columns = file2.columns.str.lower()
```

```python
file3.columns = file3.columns.str.lower()
```

```python
#2.renaming columns
file1.rename(columns={'customer' : 'customer id', 'st' : 'state','customer lifetime value' : 'customer ltv'},inplace = True)
```

```python
file2.rename(columns={'customer': 'customer id', 'st' : 'state', 'customer lifetime value' : 'customer ltv'},inplace = True)
```

```python
#3.removing spaces and adding underscore(_). 
file1.columns = file1.columns.map(lambda x:x.replace("-","_").replace(" ","_"))

file1.head()
```

```python
file2.columns = file2.columns.map(lambda x:x.replace(" ","_"))

file2.head()
```

```python
file3.columns = file3.columns.map(lambda x:x.replace("-","_").replace(" ","_"))

file3.head()
```

```python
#Rearrange the columns in DataFrame as needed
file1 = file1.reindex(columns=['customer_id','gender','income','state','education','monthly_premium_auto','vehicle_class','policy_type','customer_ltv','number_of_open_complaints','total_claim_amount'])

file1.head()
```

```python
file2 = file2.reindex(columns=['customer_id','gender','income','state','education','monthly_premium_auto','vehicle_class','policy_type','customer_ltv','number_of_open_complaints','total_claim_amount'])
```

```python
file3= file3.reindex(columns=['customer_id','gender','income','state','education','monthly_premium_auto','vehicle_class','policy_type','customer_ltv','number_of_open_complaints','total_claim_amount'])
```

```python
#Concatenate the three dataframes
df = pd.concat([file1,file2,file3],axis = 0)
```

```python
#Which columns are numerical? Which columns are categorical?
df.dtypes
```

```python
#income and monthly_premium_auto and total_claim_amount are numerical while rest are categorical. 
```

```python
#Delete the column education and the number of open complaints from the dataframe.
df = df.drop(['education','number_of_open_complaints'],axis = 1)
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
df.info()
```

```python
#converting dataFrame of customer_ltv to numeric from categorical
df['customer_ltv'] = pd.to_numeric(df['customer_ltv'],errors='coerce')
```

```python
#multiply customer_ltv with 100
df['customer_ltv'] = df['customer_ltv']*100

df.tail(100)
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)

df.tail(100)
```

```python
#Check for duplicate rows in the data and remove if any.
#checking for duplicated rows
bool_duplicates = df.duplicated()
print(bool_duplicates)
```

```python
#removing duplicated rows
df = df.drop_duplicates(keep = 0)
print(df)
```
